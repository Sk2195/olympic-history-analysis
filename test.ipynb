```python
# Query 3- Identify the sport which was played in all summer olympics.

query = """
    SELECT
        sport AS sport_name
    FROM
        athlete
    WHERE
        season = 'Summer'
    GROUP BY
        sport
    HAVING
        COUNT(DISTINCT year) = (SELECT COUNT(DISTINCT year) FROM athlete WHERE season = 'Summer');
"""

%sql $query

```

```python
# Query 4- Identify the sport which was played only once in all summer olympics
query = """
SELECT
     sport As sport_name,
     COUNT(*)
FROM athlete 
WHERE season = 'Summer'
GROUP BY sport
HAVING count(year) = 1
"""

%sql $query
```

```python
# Query 5- Identify the sport which was played only once in all summer olympics
query = """
with cte_gold_age as (
select ID, Name, Sex, Age,
DENSE_RANK() over( order by Age desc) as rnk
from athlete
where Medal = 'Gold' )
select cte_gold_age.ID, cte_gold_age.Name, cte_gold_age.Sex, cte_gold_age.Age 
from cte_gold_age where rnk = 1
"""

%sql $query
```

```python
# Query 6 Find the Ratio of male and female athletes participated in all olympic games
query = """
SELECT
    SUM(CASE WHEN sex = 'M' THEN 1 ELSE 0 END) AS male_count,
    SUM(CASE WHEN sex = 'F' THEN 1 ELSE 0 END) AS female_count,
    (SUM(CASE WHEN sex = 'M' THEN 1 ELSE 0 END)::float / SUM(CASE WHEN sex = 'F' THEN 1 ELSE 0 END)) AS male_to_female_ratio
FROM
    athlete;
"""

%sql $query
```

```python
# Query 7 Fetch the top 5 athletes who have won the most models
query = """
select Name, count( Medal) as medal_cnt
from athlete
where Medal != 'NA'
group by Name
order by 2 desc
LIMIT 5;
"""

%sql $query
```

```python
# Query 8 Number of gold, silver, and bronze medal
query = """
with cte_gold as (
                  select  Team,  count(Medal) as gold_medal_cnt
                  from athlete where Medal = 'Gold'
                  group by Team   ),

cte_silver as (
               select  Team,  count(Medal) as silver_medal_cnt from athlete
               where Medal = 'Silver'
               group by Team  ),

cte_bronze as (
              select  Team,  count(Medal) as bronze_medal_cnt from athlete
              where Medal = 'Bronze'
              group by Team   )
SELECT 
      a.team,
      a.gold_medal_cnt as gold_count,
      b.silver_medal_cnt as silver_medal_count,
      c.bronze_medal_cnt as bronze_count
FROM cte_gold as a 
INNER JOIN cte_silver as b 
ON a.team = b.team
INNER JOIN cte_bronze c
ON a.team = c.team
ORDER BY 2 DESC
"""
%sql $query
```

```python
# Query 9 - List down total gold, silver and bronze medals won by each country corresponding to each olympics season

query = """
with cte_gold as (
                  select  team, season, count(medal) as gold_medal_cnt
                  from athlete where Medal = 'Gold'
                  group by team, season ),

cte_silver as (
               select  team,  count(medal) as silver_medal_cnt from athlete
               where Medal = 'Silver'
               group by team , season ),

cte_bronze as (
              select  team,  count(medal) as bronze_medal_cnt from athlete
              where Medal = 'Bronze'
              group by team, season  )
SELECT 
      a.team as country,
      a.season as season,
      a.gold_medal_cnt as gold_count,
      b.silver_medal_cnt as silver_medal_count,
      c.bronze_medal_cnt as bronze_count
FROM cte_gold as a 
INNER JOIN cte_silver as b 
ON a.team = b.team
INNER JOIN cte_bronze c
ON a.team = c.team
ORDER BY 2 DESC
"""
%sql $query
```

```python
# Query 10 Identify the country that most gold, most silver, and most bronze medals during each olympic season

query = """

WITH cte_gold AS (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY Season ORDER BY gold_medal_cnt DESC) AS r_n_gold
    FROM (
        SELECT Season, Team, COUNT(Medal) AS gold_medal_cnt      
        FROM athlete WHERE Medal = 'Gold'
        GROUP BY Season, Team
    ) AS a
), cte_silver AS (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY Season ORDER BY silver_medal_cnt DESC) AS r_n_silver
    FROM (
        SELECT Season, Team, COUNT(Medal) AS silver_medal_cnt      
        FROM athlete WHERE Medal = 'Silver'
        GROUP BY Season, Team
    ) AS b
), cte_bronze AS (
    SELECT *,
           ROW_NUMBER() OVER (PARTITION BY Season ORDER BY bronze_medal_cnt DESC) AS r_n_bronze
    FROM (
        SELECT Season, Team, COUNT(Medal) AS bronze_medal_cnt      
        FROM athlete WHERE Medal = 'Bronze'
        GROUP BY Season, Team
    ) AS c
)
SELECT p.Season,
       CONCAT(p.Team, ' - ', p.gold_medal_cnt) AS max_gold,
       CONCAT(q.Team, ' - ', q.silver_medal_cnt) AS max_silver,
       CONCAT(r.Team, ' - ', r.bronze_medal_cnt) AS max_bronze
FROM cte_gold AS p
JOIN cte_silver AS q ON p.Season = q.Season AND p.r_n_gold = q.r_n_silver
JOIN cte_bronze AS r ON p.Season = r.Season AND p.r_n_gold = r.r_n_bronze
WHERE p.r_n_gold = 1;

"""

%sql $query
```

```python
# Query 11 Which countries have never won gold medal but have won silver or bronze medals?

query = """
SELECT DISTINCT n.region
FROM noc_regions n
WHERE n.noc NOT IN (
    SELECT DISTINCT a.noc
    FROM athlete a
    WHERE a.medal = 'Gold'
)
AND n.noc IN (
    SELECT DISTINCT a.noc
    FROM athlete a
    WHERE a.medal IN ('Silver', 'Bronze')
);

"""
%sql $query
```

```python
# Query 12 - In which sport/event U.S. has won the highest medals?
query = """
SELECT a.sport, a.event, COUNT(a.medal) AS total_medals
FROM athlete a 
INNER JOIN noc_regions nr ON a.noc = nr.noc
WHERE nr.region = 'USA'
GROUP BY a.sport, a.event
ORDER BY total_medals DESC
LIMIT 1;
"""
%sql $query


```

```python
# Query 13 -Breaking down the ice hockey
query = """
SELECT a.event, a.sport, COUNT(a.medal) AS medal_cnt 
FROM athlete a 
INNER JOIN noc_regions nr ON a.noc = nr.noc
WHERE nr.region = 'USA' and a.medal != 'NA' AND a.sport = 'Ice Hockey'
GROUP BY a.event, a.sport 
"""
%sql $query
```
